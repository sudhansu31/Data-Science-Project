## Using Regularization technique

```python
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
lassocv=LassoCV(alphas=None,max_iter=100000,normalize=True)
```

```python
lassocv.fit(x_train,y_train)
```

```python
lassocv.score(x_test,y_test)
```

```python
lassocv.alpha_
```

```python
alpha=lassocv.alpha_
```

```python
lasso_reg_model=Lasso(alpha,max_iter=10000)
```

```python
lasso_reg_model.fit(x_train,y_train)
```

```python
lasso_reg_model.score(x_test,y_test)
```

```python
lasso_reg_model.coef_
```

**we can see almost all features are important here except genorisity which is 0.39 but i have decided not to drop becuase the coefficient is not too bad.So decided to go forward with all features**

```python
ridgecv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
ridgecv.fit(x_train,y_train)
ridgecv.alpha_
```

```python
ridge_model=Ridge(alpha=ridgecv.alpha_)
```

```python
ridge_model.fit(x_train,y_train)
```

```python
y_pred=ridge_model.predict(x_test)
```

```python
ridge_model.score(x_test,y_test)
```
